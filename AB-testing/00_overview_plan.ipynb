## Цели

<div style="padding: 1.5em; background: #ffffff; border-radius: 8px; margin: 1em 0;">
  <ul style="margin: 0; padding: 0; list-style: none;">
    <li style="margin: 0.9em 0; padding-left: 2em; position: relative; line-height: 1.6; color: #212529;">
      <span style="position: absolute; left: 0; top: 0.2em; width: 6px; height: 6px; background: #667eea; border-radius: 50%;"></span>
      Сформировать практическое понимание, когда и как использовать A/B‑тесты.
    </li>
    <li style="margin: 0.9em 0; padding-left: 2em; position: relative; line-height: 1.6; color: #212529;">
      <span style="position: absolute; left: 0; top: 0.2em; width: 6px; height: 6px; background: #667eea; border-radius: 50%;"></span>
      Обеспечить статистическую корректность: рандомизация, мощность, множественные проверки, sequential‑контроль.
    </li>
    <li style="margin: 0.9em 0; padding-left: 2em; position: relative; line-height: 1.6; color: #212529;">
      <span style="position: absolute; left: 0; top: 0.2em; width: 6px; height: 6px; background: #667eea; border-radius: 50%;"></span>
      Научить проектировать метрики и снижать дисперсию (ковариаты, CUPED/CUPAC).
    </li>
    <li style="margin: 0.9em 0; padding-left: 2em; position: relative; line-height: 1.6; color: #212529;">
      <span style="position: absolute; left: 0; top: 0.2em; width: 6px; height: 6px; background: #667eea; border-radius: 50%;"></span>
      Освоить частотные и байесовские подходы, а также бутстрэп и непараметрику.
    </li>
    <li style="margin: 0.9em 0; padding-left: 2em; position: relative; line-height: 1.6; color: #212529;">
      <span style="position: absolute; left: 0; top: 0.2em; width: 6px; height: 6px; background: #667eea; border-radius: 50%;"></span>
      Понять сложные дизайны (switchback, GEO, факторные) и ограничения данных (heavy‑tails, цензурирование, лаги).
    </li>
    <li style="margin: 0.9em 0; padding-left: 2em; position: relative; line-height: 1.6; color: #212529;">
      <span style="position: absolute; left: 0; top: 0.2em; width: 6px; height: 6px; background: #667eea; border-radius: 50%;"></span>
      Разобраться в инженерных аспектах платформы экспериментов и продуктовых trade‑offs.
    </li>
    <li style="margin: 0.9em 0; padding-left: 2em; position: relative; line-height: 1.6; color: #212529;">
      <span style="position: absolute; left: 0; top: 0.2em; width: 6px; height: 6px; background: #667eea; border-radius: 50%;"></span>
      Уметь интерпретировать гетерогенность эффектов и строить uplift‑модели.
    </li>
  </ul>
</div>


## Структура (дорожная карта)

  <ul style="margin: 0; padding: 0; list-style: none;">
    <li style="margin: 0.75em 0; padding: 0.6em 1em; padding-left: 2.5em; position: relative; line-height: 1.6; background: #f8f9fa; border-radius: 6px; border-left: 3px solid #667eea; transition: all 0.2s;">
      <span style="position: absolute; left: 1em; top: 0.75em; color: #667eea; font-weight: 600;">1</span>
      <strong style="color: #1a1a1a; font-weight: 600;">Основы:</strong> <span style="color: #2c3e50;">определения A/B, рандомизация, A/A‑проверки баланса</span>
    </li>
    <li style="margin: 0.75em 0; padding: 0.6em 1em; padding-left: 2.5em; position: relative; line-height: 1.6; background: #f8f9fa; border-radius: 6px; border-left: 3px solid #667eea;">

    </li>

      <span style="position: absolute; left: 1em; top: 0.75em; color: #667eea; font-weight: 600;">4</span>

      <span style="position: absolute; left: 1em; top: 0.75em; color: #667eea; font-weight: 600;">6</span>
